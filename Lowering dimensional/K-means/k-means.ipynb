K-Means

```python
import sklearn
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.cluster import KMeans
```

```python
from sklearn import datasets
```

```python
# Download Fisher's iris DS
iris = datasets.load_iris()
```

```python
# Split the DS
X = iris.data
Y = iris.target
```

```python
# Split our DS on 3 clusters with k-means method
k_means3 = KMeans(n_clusters = 3).fit(X)
```

```python
# Show the clusters centers
k_means3.cluster_centers_
```

```python
fig = plt.figure(figsize=(10, 4))
ax = fig.add_subplot(121)
plt.scatter(X[:, 0], X[:, 1], c=k_means3.labels_, marker='o', cmap='rainbow')
ax = fig.add_subplot(122)
plt.scatter(X[:, 0], X[:, 1], c=Y, marker='o', cmap='rainbow')
```

```python
fig = plt.figure(figsize=(10, 4))
ax = fig.add_subplot(121)
plt.scatter(X[:, 0], X[:, 2], c=k_means3.labels_, marker='o', cmap='rainbow')
ax = fig.add_subplot(122)
plt.scatter(X[:, 0], X[:, 2], c=Y, marker='o', cmap='rainbow')
```
