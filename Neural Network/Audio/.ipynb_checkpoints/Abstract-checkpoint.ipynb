librosa is a python package for music and audio analysis. It provides the building blocks necessary to create music information retrieval systems.
https://librosa.org/doc/latest/index.html

У аудиофайла есть признаки:
* Ритм - список музыкальных нот
* Темп - скорость с которой повторяются паттерны в аудиофайле
* Бит - отрезок времени, ритм который выстукивается в песне
* Такт - логичное деление битов, обычно содержит 3-4 бита
* Интервал - 8 нот обычно

Признаки аудиофайла:
* Спектральный центроид
* Спектральный спад частоты
* Мел-частотные кепстральные коэффициенты
* Частота цветности

Чтобы лучше подать аудиосигнал в модель необходимо разделить аудиодорожку частями, провести стандартизацию и подавать их по одной

## Признаки:

* Zero crossing rate - частота пересечения нуля - частота изменения знака сигнала. Используется в основном для сегментации речевого порога, а также для определения звуков с высокочастотным шумом, для определения следов монтажа;

* Spectr features - Спектограмма - изображение показывающее зависимость спектральной плотности мощности сигнала от времени. Для идентификации речи, звуков животных, в гидролокации, и тд. Зависимость частоты от времени, иногда применяется трехмерная визуализация с амплитуда;

* Spectric cenroid - Спектральный центроид - Указывает, на какой частоте сосредоточена энергия спектра или, другими словами, указывает, где расположен “центр масс” для звука, и рассчитывается как средневзвешенное значение всех частот. Мера для характеристики спектра. Эффективно указывают зашумленность сигнала. Позволяет оценивать особенности звука; https://librosa.org/doc/main/generated/librosa.feature.spectral_centroid.html

* Mel - мел - единица высоты звука, основана на восприятии этого звука нашими органами слуха;

* Chroma features - Цветность высоты тона - представление для музыкального звука, при котором весь спектр проецируется на 12 контейнеров, представляющих 12 различных полутонов музыкальной актавы. Передают гармонические хар-ки музыка и усточивы к изменению тембра и инструментов;


* Добавление перед блоком полносвязанного классификатора блока LSTM;
* Добавление блока свертки;
* Добавление двух блоков одновременно;

## Варианты подачи данных в модель:

* (! Эффективнее !) Подача данных о параметрах звука и его признаками на различные входные слои;
* Подача данных о параметрх звука вместе с его признаками на один входной слой;


## Нюансы при создании модели:

* Использование свертки только с использованием классификатора;
* Чем чаще происходят колебания - тем звук выше;
* Активатор Elu - чаще показывает эффективность в задачах анализа аудиосигнала;

Одномерная свертка ConvID может обработать:
* Спектральный цетроид;
* Спектральный спад частоты;
* Ширина полосы частот;

Не может обработать:
* Частота цветности;
* Мел кепстральные коэффициенты;

# Применение:

* Голосовые помощники;
* Распознование речи в текст;
* Классификация аудио;
* Обнаружение аномалий в звуковых дорожках;
* Машинное творчество - композиции;

# Трудности:

* Акценты и шум;
* Семантические ошибки;
* Один канал, много говорящих;
* Контекст;
