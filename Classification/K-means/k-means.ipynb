```python
from sklearn.cluster import KMeans
```

```python
from sklearn import datasets
```

```python
# Скачивание ДС с ирисами Фишера
    # 1. Длина наружной доли околоцветника (sepal length)
    # 1. Ширина наружной доли околоцветника (sepal width)
    # 1. Длина внутренней доли околоцветника (petal length)
    # 1. Ширина внутренней доли околоцветника (petal width)
iris = datasets.load_iris()
```

```python
X = iris.data
Y = iris.target
```

```python
k_means3 = KMeans(n_clusters = 3).fit(X)
```

```python
k_means3.cluster_centers_
```

```python
fig = plt.figure(figsize=(10, 4))
ax = fig.add_subplot(121)
plt.scatter(X[:, 0], X[:, 1], c=k_means3.labels_, marker='o', cmap='rainbow')
ax = fig.add_subplot(122)
plt.scatter(X[:, 0], X[:, 1], c=Y, marker='o', cmap='rainbow')
```

```python
fig = plt.figure(figsize=(10, 4))
ax = fig.add_subplot(121)
plt.scatter(X[:, 0], X[:, 2], c=k_means3.labels_, marker='o', cmap='rainbow')
ax = fig.add_subplot(122)
plt.scatter(X[:, 0], X[:, 2], c=Y, marker='o', cmap='rainbow')
```
